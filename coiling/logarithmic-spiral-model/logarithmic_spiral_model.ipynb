# 対数らせんモデル
極座標を考え，動径$r$，偏角$\theta$について以下の関係があるとする．
$$
\label{eq:ls}
\frac{dr}{d\theta} = \alpha \theta \tag{1}
$$
$\alpha$は成長率を表すパラメータ．
初期値$r(0) = r_0$として解くと，
$$
r(\theta) = r_0 e^{\alpha \theta}
$$
となる．

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
def logSpiral(a, r0, theta):
    """対数螺旋
    
    対数螺旋の座標値を返す関数
    
    Args:
        a: 対数螺旋の拡大率
        r0: 動径の初期値
        theta: 回転角
        
    Returns:
        x, y: 対数螺旋上の座標値
    
    """
    r = r0*np.exp(a*theta)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    return (x,y)
```

```python
r0 = 1
a = 0.2

theta = np.linspace(0, 8*np.pi,1000)

x, y = logSpiral(a, r0, theta)
```

```python
plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal')
plt.plot(x,y)
```

```python
cases_a = [0.01,0.1,0.2,0.4,0.8,1.6]
fig, axs = plt.subplots(2, 3, figsize=(12,6), constrained_layout=True)
for ax, a in zip(axs.flat[:], cases_a):
    r0 = 1
    theta = np.linspace(0, 8*np.pi,1000)
    x, y = logSpiral(a, r0, theta)
    
    ax.set_title(r'$\alpha$ = %s' % str(a))
    ax.set_aspect('equal')
    ax.plot(x, y)
```

## 黄金長方形と黄金らせん
辺の長さの比が黄金比（$1:\phi$，ただし$\phi = \frac{1+\sqrt{5}}{2}$）になっている長方形を黄金長方形という．
ある黄金長方形に直線を一本引いて，この長方形の短辺と同じ辺の長さをもつ正方形を切り出すと，残った長方形もまた黄金長方形になる．

（あるいは逆に，黄金長方形の長辺と同じ辺の長さをもつ正方形を“付け加える”と新たにできた大きな長方形はまた黄金長方形になる．）

このプロセスを繰り返し，（“外側”を常に正方形に割り当てれば）分割した線と長方形の“外側”の辺との交点は，ある対数らせんの上に配置される．
この対数らせんは黄金らせんと呼ばれている．

具体的には，対数らせんモデルにおいて$\alpha = \frac{2\log{\phi}}{\pi}$のとき，黄金らせんとなる．

参考：[Golden Rectangle | Wolfram MathWorld](http://mathworld.wolfram.com/GoldenRectangle.html)

```python
r0 = 1
a = np.log((1+np.sqrt(5))/2)/(np.pi/2) # 黄金らせん

theta = np.linspace(0, 8*np.pi,1000)

x, y = logSpiral(a, r0, theta)
```

```python
plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal')
plt.plot(x,y)
```
