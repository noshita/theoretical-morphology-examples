# 対数螺旋
$$
\frac{dr}{d\theta} = a \theta
$$

$$
r(\theta) = r_0 e^{a \theta}
$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logSpiral(a, r0, theta):
    """対数螺旋
    
    対数螺旋の座標値を返す関数
    
    Args:
        a: 対数螺旋の拡大率
        r0: 動径の初期値
        theta: 回転角
        
    Returns:
        x, y: 対数螺旋上の座標値
    
    """
    r = r0*np.exp(a*theta)
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    
    return (x,y)
```

```python
r0 = 1
a = 0.2

theta = np.linspace(0, 8*np.pi,1000)

x, y = logSpiral(a, r0, theta)
```

```python
plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal')
plt.plot(x,y)
```

```python
plt.figure(figsize=(7,7))
g = plt.subplot()
g.set_aspect('equal')
g.plot(x,y)
```
